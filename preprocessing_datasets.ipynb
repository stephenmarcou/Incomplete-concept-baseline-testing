```python
import pickle 
import re
import os

DATA_DIR = 'Data/'
```

### Preprocess dataset from CBM repo such that we point to correct image files

```python
def change_img_path(old_file_path, new_file_path):

    with open(DATA_DIR + old_file_path, "rb") as f:
        data = pickle.load(f)
    
    
    # Change image path
    for i in range(len(data)):
        image_metadata = data[i]
        old_img_path = image_metadata["img_path"]
        rel_part_old_path = re.search(r"/datasets/(.*)", old_img_path).group(1)
        new_img_path = DATA_DIR + "CUB_200_2011/" + rel_part_old_path
        data[i]["img_path"] = new_img_path
        
    new_dir = new_file_path.split("/")[0]
    if not os.path.exists(DATA_DIR + new_dir):
        os.makedirs(DATA_DIR + new_dir)

        print(f"File {DATA_DIR + new_file_path} already exists. Please choose a different name or delete the existing file.")
        return

    # Save new pickle file
    with open(DATA_DIR + new_file_path, "wb") as f:
        pickle.dump(data, f)
        
        
change_img_path("class_attr_data_10_from_CBM_repo/train.pkl", "class_attr_data_10_updated/train.pkl")
change_img_path("class_attr_data_10_from_CBM_repo/val.pkl", "class_attr_data_10_updated/val.pkl")
change_img_path("class_attr_data_10_from_CBM_repo/test.pkl", "class_attr_data_10_updated/test.pkl")

        

    
    
```

### Random concept removal 

```python
import random
```

```python
data = pickle.load(open(DATA_DIR + "class_attr_data_10_updated/train.pkl", "rb"))
print(data[0].keys())
print(data[0]["class_label"])
```

```python

N_ATTR = 112
KEEP_FRACTION = 0.5
N_KEEP = int(N_ATTR * KEEP_FRACTION)


n_attr_idx = list(range(N_ATTR))
n_attr_idx_keep = random.sample(n_attr_idx, N_KEEP)

def random_concept_removal(new_file_path, old_file_path):
    with open(DATA_DIR + old_file_path, "rb") as f:
        data = pickle.load(f)
    
    for i in range(len(data)):
        attribute_label = data[i]["attribute_label"]
        attribute_certainty = data[i]["attribute_certainty"]
        new_attribute_label = [attribute_label[j] for j in n_attr_idx_keep]
        new_attribute_certainty = [attribute_certainty[j] for j in n_attr_idx_keep]
        
        data[i]["attribute_label"] = new_attribute_label
        data[i]["attribute_certainty"] = new_attribute_certainty

    
    new_dir = new_file_path.split("/")[0]
    if not os.path.exists(DATA_DIR + new_dir):
        os.makedirs(DATA_DIR + new_dir)
    
    
    with open(DATA_DIR + new_file_path, "wb") as f:
        pickle.dump(data, f)    



random_concept_removal("class_attr_data_10_updated_random_removed_concepts/train.pkl", "class_attr_data_10_updated/train.pkl")
random_concept_removal("class_attr_data_10_updated_random_removed_concepts/val.pkl", "class_attr_data_10_updated/val.pkl")
random_concept_removal("class_attr_data_10_updated_random_removed_concepts/test.pkl", "class_attr_data_10_updated/test.pkl")
```
